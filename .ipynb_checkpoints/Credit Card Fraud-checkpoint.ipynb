As we saw this is a big imbalanced dataset, there are so many techniques to trate with imbalance:

- Oversampling: Oversampling solves the imbalance by replicating data from the minority class.

- Undersampling: It is based on reducing the number of observations of the majority class to achieve a balanced database.It is advisable to use this method only when our database is very large and has many samples, since reducing instances also improves the model training time.

- Undersampling & Oversampling: Using Undersampling followed by Oversampling helps us to find a ratio that allows us not to lose some concepts of the majority class, and not to have to generate as many copies of the minority class, thus reducing overfitting.

We have to treat that because if we use an imbalanced dataset our clasifiers will asume that in most cases there are no frauds! What we want for our model is to be certain when a fraud occurs. And with an imbalanced dataset we can not see easily the correlation between the features and the class, so we need to deal this imbalance.

#### Random Oversampling 

Random Oversampling is based on randomly duplicating instances of the minority class and adding them to the dataset. The instances are selected from the training set with replacement, i.e. the same instance can be selected several times. With Oversampling we can indicate the proportion of the majority class we want to keep.

```python
ROS = RandomOverSampler(random_state=42)
X_res, y_res = ROS.fit_resample(Org_X_train, Org_y_train)
```

Create a class for the outliers extraction and include it in the pipeline

```python


class ResampleAndRemoveOutliers(BaseEstimator):
    def __init__(self, sampler=RandomOverSampler(random_state=42)):
        self.sampler = sampler

    def fit(self, X, y=None):
        # Fit method is required by scikit-learn convention, but the actual work is done in fit_resample.
        # This is to ensure compatibility with imblearn's pipeline steps that expect a fit_resample method.
        return self

    def fit_resample(self, X, y):
        # Perform the resampling first
        X_resampled, y_resampled = self.sampler.fit_resample(X, y)
        print("Start Resample")
        
        columns_df = pd.concat([X_resampled, y_resampled], axis=1)
        correlations_with_class = new_df.corrwith(new_df['Class']).sort_values(ascending=False)
        top_positive = correlations_with_class.head(4).index.tolist()
        top_negative = correlations_with_class.tail(3).index.tolist()
        selected_columns = top_positive + top_negative
        # Then, perform the outlier removal
        selected_columns = [col for col in selected_columns if col in X_resampled.columns]
        print("Selected Columns: ",selected_columns)
        for col in selected_columns:
            # Calculate the IQR for each column and filter outliers
            q25, q75 = np.percentile(X_resampled[col], [25, 75])
            iqr = q75 - q25
            cut_off = iqr * 1.5
            lower, upper = q25 - cut_off, q75 + cut_off

            # Determine which rows to keep
            keep_rows = (X_resampled[col] >= lower) & (X_resampled[col] <= upper)
            X_resampled = X_resampled[keep_rows]
            y_resampled = y_resampled[keep_rows]
        print("Resample Done")
        return X_resampled, y_resampled
```

#### Random Undersampling

Random Undersampling is based on randomly selecting instances of the majority class and eliminating them from the training set, reducing the number of instances of the majority class in the new version of the training set, this is a sampling technique without replacement. Where the objective of sampling is to create a training dataset by modifying the distribution of the target variable, obtaining a more balanced dataset. With Undersampling we can indicate the proportion of the majority class we want to keep.

```python
from imblearn.under_sampling import RandomUnderSampler

print('Resampled dataset shape %s' % Counter(Org_y_train))
rus = RandomUnderSampler(random_state=42)
RUS_X_train, RUS_y_train = rus.fit_resample(Org_X_train, Org_y_train)
print('Resampled dataset shape %s' % Counter(RUS_y_train))
```
